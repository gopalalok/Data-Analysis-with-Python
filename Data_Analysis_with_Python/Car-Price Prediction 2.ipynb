### Convert categorical code into order

```python
df['fuelsystem'].unique()
```

```python
from pandas.api.types import CategoricalDtype
df['fueltype'] = df['fueltype'].astype(CategoricalDtype(categories=['gas', 'diesel'], ordered = True)).cat.codes
```

```python
df['fueltype'].unique()
```

```python
df['aspiration'] = df['aspiration'].astype(CategoricalDtype(categories=['std', 'turbo'], ordered = True)).cat.codes
df['doornumber'] = df['doornumber'].astype(CategoricalDtype(categories=['two', 'four'], ordered = True)).cat.codes
df['carbody'] = df['carbody'].astype(CategoricalDtype(categories=['convertible', 'hatchback', 'sedan', 'wagon', 'hardtop'], ordered = True)).cat.codes
df['drivewheel'] = df['drivewheel'].astype(CategoricalDtype(categories=['rwd', 'fwd', '4wd'], ordered = True)).cat.codes
df['enginelocation'] = df['enginelocation'].astype(CategoricalDtype(categories=['front', 'rear'], ordered = True)).cat.codes
df['enginetype'] = df['enginetype'].astype(CategoricalDtype(categories=['dohc', 'ohcv', 'ohc', 'l', 'rotor', 'ohcf', 'dohcv'], ordered = True)).cat.codes
df['cylindernumber'] = df['cylindernumber'].astype(CategoricalDtype(categories=['four', 'six', 'five', 'three', 'twelve', 'two', 'eight'], ordered = True)).cat.codes
df['fuelsystem'] = df['fuelsystem'].astype(CategoricalDtype(categories=['mpfi', '2bbl', 'mfi', '1bbl', 'spfi', '4bbl', 'idi', 'spdi'], ordered = True)).cat.codes
```

```python
data = df.drop(['car_ID','CarName','price'],axis = 1).columns
X = df[data].values
```

```python
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

### Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score

def test_model(model, X_train=X_train, y_train=y_train):
    cv = KFold(n_splits = 3, shuffle=True, random_state = 45)
    r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model, X_train, y_train, cv=cv, scoring = r2)
    score = [r2_val_score.mean()]
    return score
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
test_model(lr)
```

```python
import sklearn.linear_model as linear_model
rdg = linear_model.Ridge()
test_model(rdg)
```

```python
lasso = linear_model.Lasso(alpha=1e-4)
test_model(lasso)
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state=21)
test_model(dt_reg)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators = 1000, random_state=51)
test_model(rf_reg)
```

### Bagging & boosting

```python
from sklearn.ensemble import BaggingRegressor, GradientBoostingRegressor
br_reg = BaggingRegressor(n_estimators=1000, random_state=51)
gbr_reg = GradientBoostingRegressor(n_estimators=1000, learning_rate=0.1, loss='ls', random_state=51)
```

```python
test_model(br_reg)
```

```python
test_model(gbr_reg)
```

### XGBoost

```python
import xgboost
#xgb_reg=xgboost.XGBRegressor()
xgb_reg = xgboost.XGBRegressor(bbooster='gbtree', random_state=51)
test_model(xgb_reg)
```
